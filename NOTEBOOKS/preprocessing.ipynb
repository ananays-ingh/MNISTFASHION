```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from tensorflow.keras.utils import to_categorical 
```

```python
data_train= pd.read_csv('/Users/mr.engineer/Desktop/Code/DSPROJECT/MNIST_FASHION/DATA/fashion-mnist_train.csv')
data_test = pd.read_csv('/Users/mr.engineer/Desktop/Code/DSPROJECT/MNIST_FASHION/DATA/fashion-mnist_test.csv')

X_train = data_train.drop(columns='label')
y_train = data_train['label'].values
X_test = data_test.drop(columns='label')
y_test = data_test['label'].values
sample = X_train.sample(1).values.reshape(28,28)

sns.heatmap(sample)
```

```python
X_train = X_train/255.0 # Normalizing
y_train_cat = to_categorical(y_train,num_classes=10) # OHE
y_test_cat = to_categorical(y_test,num_classes=10)
X_test = X_test/255.0
```

```python
df_y_train_cat = pd.DataFrame(y_train_cat)
df_y_test_cat = pd.DataFrame(y_test_cat)

df_X_train = pd.DataFrame(X_train)
df_X_test = pd.DataFrame(X_test)


path = '/Users/mr.engineer/Desktop/Code/DSPROJECT/MNIST_FASHION/DATA/PROCESSED/'

df_X_train.to_csv(path+"X_train_processed.csv", index=False)
df_y_train_cat.to_csv(path+"y_train_cat.csv", index=False)
df_X_test.to_csv(path+"X_test_processed.csv", index=False)
df_y_test_cat.to_csv(path+"y_test_cat.csv", index=False)
```

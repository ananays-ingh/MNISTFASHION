```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data_train = pd.read_csv("/Users/mr.engineer/Desktop/Code/DSPROJECT/MNIST_FASHION/DATA/fashion-mnist_train.csv")
data_test= pd.read_csv("/Users/mr.engineer/Desktop/Code/DSPROJECT/MNIST_FASHION/DATA/fashion-mnist_test.csv")

X_train = data_train.drop(columns='label')
y_train = data_train['label'].values
X_test= data_test.drop(columns='label')
y_test = data_test['label'].values
```

```python
print('X_train shape',X_train.shape)
print('y_train shape',y_train.shape)
print('X_test shape',X_test.shape)
print('y_test shape',y_test.shape)

print('Unique Labels',np.unique(y_test))
print('No of classes',len(np.unique(y_test)))
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x=y_train)
plt.title('Class Distribution')
plt.grid(True)
```

```python
classes = ['t-shirt', 'trouser', 'pullover','dress', 'coat', "sandal", "shirt", 'sneaker', "Bag", "ankle boot"]

plt.figure(figsize=(15, 7))
for i in range(10):
    current_class = X_train[y_train==i]
    img = current_class.sample(1).values.reshape(28,28)
    plt.subplot(2,5,i+1)
    plt.imshow(img,cmap='gray')
    plt.title(str(i)+' th class: '+str(classes[i]))
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(X_train.values.flatten(), bins = 50, kde = True,)
plt.title("pixel Intensity Distribution ")
plt.show()
```
